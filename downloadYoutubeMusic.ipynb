```
import yt_dlp
import ffmpeg

def download_playlist_info(playlist_url):
    ydl = {
        'quiet': True,
        'extract_flat': True,  # This ensures we only extract metadata, not download
        'skip_download': True,
    }

    with yt_dlp.YoutubeDL(ydl) as ydl:
        playlist_info = ydl.extract_info(playlist_url, download=False)
    return playlist_info

def get_video_urls(playlist_info):
    video_urls = []
    for entry in playlist_info.get('entries', []):
        video_urls.append(entry['url'])
    return video_urls

def get_video_filename(video_url):
    ydl_opts = {
        'quiet': True,
        'no_warnings': True,
        'writethumbnail': True,
        'skip_download': True
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(video_url, download=False)
        if 'creator' in info and 'title' in info:
            filename = info['creator'] + '-' + info['title']
        else:
            filename = info['title']
        return filename

def downloadYoutubeVideo(filename, url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': filename + '.%(ext)s',
        'extractaudio': True,
        'writethumbnail': True,  # 下載縮圖
        'add-metadata': True,   # 加入詮釋資料（包含作者資訊）
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

def convertWebmToMp3(input_file, output_file, image_file):
    # 提取音频
    audio_stream = ffmpeg.input(input_file).audio

    # 创建带封面的mp3文件
    stream = ffmpeg.output(audio_stream, output_file, **{'metadata:s:v': 'title=Album cover',
                              'metadata:s:v': 'comment=Cover (front)',
                              'codec:v': 'mjpeg', 'map': '0:a', 'map': '1:v'},
                           shortest=None, **{'v': 'error'},
                           **{'ab': '320k'})

    # 添加封面图片
    stream = ffmpeg.input(image_file, **{'loop': 1}).output(stream, **{'c:v': 'mjpeg'})

    # 执行转换
    ffmpeg.run(stream, overwrite_output=True)
```

```
if __name__ == "__main__":
    playlist_url = 'https://music.youtube.com/playlist?list=PL5FgdjISrY7wfDHo_VZv63lIPqXRl4Bn7'
    playlist_info = download_playlist_info(playlist_url)

    if playlist_info:
        video_urls = get_video_urls(playlist_info)
        for url in video_urls:
            filename = get_video_filename(url)
            print(filename, url)
            downloadYoutubeVideo(filename, url)
            convertWebmToMp3ToWav(filename+'.webm', filename+'.mp3', filename+'.webp')
    else:
        print("Failed to extract playlist information.")
```
